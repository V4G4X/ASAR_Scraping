```python
import wget
import ssl
import os
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# Enter NCRB Individual PDF url
pdf_url = 'https://ncrb.gov.in/sites/default/files/adsi_reports_previous_year/Table-1.0_2019.pdf'
fileName = pdf_url[67:]
try:
    #Delete the file to redownload it
    os.remove(fileName)
except FileNotFoundError as e:
    #Ignore in case FileNotFound Error occurs
    pass
# Download report
fileName = wget.download(pdf_url,fileName)
```

```python
import numpy as np
import pandas as pd
import camelot
```

```python
fileName
```

```python
tables = camelot.read_pdf(fileName,
    flavor='stream',
    table_areas=[
        "56.75437802907916,671.7985151700498,547.3430694668822,82.11931584024649"
        ],
    columns=[
        "74,248.18016155088856,294.3532148626818,361.6889176090469,417.4813570274637,492.5125686591277"
        ],
    row_tol=10,
    split_text=True,
    strip_text='\n',
    pages="all"
)
```

```python
def printnAllTables(tables):
    for i in range(len(tables)):
        print("Page No: ",(i+1))
        print(tables[i].df)
        print()
```
